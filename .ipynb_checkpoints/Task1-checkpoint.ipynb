Regularised L2 regression

Implement using matrix multiplication: 

$ \hat{w} = (X^T X + \gamma I) X^Ty$

$ \hat{y} = \hat{w} X$

```python
gamma = 2.3
# get weights 
shape = X_train.shape
print(shape)
w = np.linalg.inv(X_train.T @ X_train + gamma * np.identity(shape[1])) @ X_train.T @ y_train
print(w.shape)
# use it to estimate results
y_hat_train = X_train @ w
y_hat_test  = X_test @ w
```

```python
from sklearn.metrics import mean_squared_error
import math

def FindOutliers(yActual, yPredicted, outlierThreshold, testOrTrain):
    """Takes actual and predicted values to calculate RMSE and return outlier indices.
    An outlier threshold is supplied, where values with absolute residuals greater than
    outerlierThreshold are identified as outliers."""
    MSE = mean_squared_error(yActual, yPredicted)

    RMSE = math.sqrt(MSE)
    print(f"RMSE for {testOrTrain} data: {RMSE}")
    
    residuals = []
    for i in range(len(yActual)):
        residual = yActual[i]-yPredicted[i]
        residuals.append(residual)
    
    outliers = [resid for resid in residuals if abs(resid) > outlierThreshold]
    outlierIndices = [residuals.index(outlier) for outlier in outliers]

    yActualValsOutliers = [yActual[outlierIndex] for outlierIndex in outlierIndices]
    yPredictedValsOutliers = [yPredicted[outlierIndex] for outlierIndex in outlierIndices]
    
    
    plt.figure(figsize=(8,8))
    plt.scatter(yActual, yPredicted, s=3)
    plt.xlabel('Target Values')
    plt.ylabel('Estimated Values')
    plt.title(f"{testOrTrain.capitalize()} set")
    
    plt.plot(yActual, yActual, "k-")
    
    plt.plot(yActualValsOutliers, yPredictedValsOutliers, "rx")
    
    return outlierIndices


outlierThreshold = 1
outlierIndices = FindOutliers(y_test, y_hat_test, outlierThreshold, "test")
FindOutliers(y_train, y_hat_train, outlierThreshold, "train")




XTestOutlierVals = []
yTestOutlierVals = []
print(len(X_test))
for i in outlierIndices:
    XTestOutlierVals.append(X_test[i])
    yTestOutlierVals.append(y_test[i])


outlierStructures = []
for i in range(len(yTestOutlierVals)):
    selectedRow = sol.loc[(sol['LogS.M.'] == yTestOutlierVals[i]) & (sol["MW"] == XTestOutlierVals[i][0])]

    outlierStructures.append(list(selectedRow["SMILES"].astype("string")))

#currently outlierStructures contains a lists of list, with each sublist having length 1.
#using list comprehension to pull the strings out of their len=1 list
print(outlierStructures)
outlierStructures = [outlierStruct[0] for outlierStruct in outlierStructures]

def GetOutlierSMILES(outlierIndices, X, y):
    XTestOutlierVals = []
    yTestOutlierVals = []
    for i in outlierIndices:
        XTestOutlierVals.append(X[i])
        yTestOutlierVals.append(y[i])


    outlierStructures = []
    for i in range(len(yTestOutlierVals)):
        selectedRow = sol.loc[(sol['LogS.M.'] == yTestOutlierVals[i]) & (sol["MW"] == XTestOutlierVals[i][0])]

        outlierStructures.append(list(selectedRow["SMILES"].astype("string")))

    outlierStructures = [outlierStruct[0] for outlierStruct in outlierStructures]
    return outlierStructures

```

### PLS Regression, aided by PCA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn import model_selection
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split
from sklearn.cross_decomposition import PLSRegression
from sklearn.metrics import mean_squared_error


sol = pd.read_excel("Husskonen_Solubility_Features.xlsx", engine='openpyxl', verbose=False)

X = sol[sol.columns[5:-1]] # getting the relevant data
X = X.to_numpy()
N, p = X.shape
y = sol["LogS.M."].values

#define cross-validation method
cv = KFold(n_splits=10)

mse = []
n = len(X)


# Calculate MSE using cross-validation, adding one component at a time
for i in np.arange(1, 22):
    pls = PLSRegression(n_components=i)
    score = -1*model_selection.cross_val_score(pls, scale(X), y, cv=cv,
               scoring='neg_mean_squared_error').mean()
    print(i)
    mse.append(score)

# plot test MSE vs. number of components
plt.figure(1)
plt.plot(mse)
plt.xlabel('Number of PLS Components')
plt.ylabel('MSE')
plt.title('hp')
plt.show()

#split the dataset into training (70%) and testing (30%) sets
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)

#calculate RMSE
pls = PLSRegression(n_components=7)
pls.fit(scale(X_train), y_train)

print(np.sqrt(mean_squared_error(y_test, pls.predict(scale(X_test)))))

PLS_yHatTest = pls.predict(scale(X_test))
PLS_yHatTrain = pls.predict(scale(X_train))
```

For clustering followed tutorial from https://projects.volkamerlab.org/teachopencadd/talktorials/T005_compound_clustering.html

```python
# Converting SMILES to RDKit objects
outliers = []
for smiles in outlierSMILES_PLS:
    try:
        outliers.append(Chem.MolFromSmiles(smiles))
    except:
        continue # Couldn't stop the RDKit error showing up.
outliers = list(filter(None, outliers))
outliers
```

```python
import mols2grid
from rdkit import Chem
from rdkit import DataStructs
from rdkit.ML.Cluster import Butina
from rdkit.Chem import Draw
from rdkit.Chem import rdFingerprintGenerator

#Visual examination of outliers
mols2grid.display(outliers)
```

```python
# Create fingerprints for all molecules
rdkit_gen = rdFingerprintGenerator.GetRDKitFPGenerator(maxPath=5)
fingerprints = [rdkit_gen.GetFingerprint(outlier) for outlier in outliers]

# Fingerprinting summary
print("Number of compounds converted:", len(fingerprints))
print("Fingerprint length per compound:", len(fingerprints[0]))
```

```python
def tanimoto_distance_matrix(fp_list):
    """Calculate distance matrix for fingerprint list"""
    dissimilarity_matrix = []
    for i in range(1, len(fp_list)):
        # Compare the current fingerprint against all the previous ones in the list
        similarities = DataStructs.BulkTanimotoSimilarity(fp_list[i], fp_list[:i])
        # Since we need a distance matrix, calculate 1-x for every element in similarity matrix
        dissimilarity_matrix.extend([1 - x for x in similarities])
    return dissimilarity_matrix
```

```python
def cluster_fingerprints(fingerprints, cutoff=0.2):
    """Cluster fingerprints
    Parameters:
        fingerprints
        cutoff: threshold for the clustering
    """
    # Calculate Tanimoto distance matrix
    distance_matrix = tanimoto_distance_matrix(fingerprints)
    # Cluster the data with the implemented Butina algorithm:
    clusters = Butina.ClusterData(distance_matrix, len(fingerprints), cutoff, isDistData=True)
    clusters = sorted(clusters, key=len, reverse=True)
    return clusters
```

```python
# Run the clustering procedure for the dataset
clusters = cluster_fingerprints(fingerprints, cutoff=0.8)

print("total # clusters: ", len(clusters))
```

```python
# Choosing cutoff for clustering
for cutoff in np.arange(0.0, 1.0, 0.2):
    clusters = cluster_fingerprints(fingerprints, cutoff=cutoff)
    fig, ax = plt.subplots(figsize=(15, 4))
    ax.set_title(f"Threshold: {cutoff:3.1f}")
    ax.set_xlabel("Cluster index")
    ax.set_ylabel("Number of molecules")
    ax.bar(range(1, len(clusters) + 1), [len(c) for c in clusters], lw=5)
    display(fig)
# I can't work out why all the graphs are repeated?
# Need a threshold of at least 0.8 to have any meaningful clusters
```
